```python
# достает актуальный список тикеров акций и капитализации компаний из индекса NASDAQ Composite

import requests  
from bs4 import BeautifulSoup
from tqdm import tqdm

def get_tickers(p):
    
    url = f'https://disfold.com/stock-index/nasdaq-composite/companies/?page={p}'
    
    response = requests.get(url)
    
    tree = BeautifulSoup(response.content, 'html')
    
    comp = tree.find_all('td', {'class' : 'center-align'})
    
    
    i = 0 # данные находятся в таблице 1 х 4, тикеры стоят на второй позиции
    x = len(comp)
    tickers = []
    
    for company in comp:
        tickers.append(comp[i].text)
        i = i + 1
        if i >x:
            break
    
    full_list = [ ]                 
    
    for ticker in tickers:
        full_list.append(ticker)
    
    return full_list
```

```python
full_list = [ ]

for p in tqdm(range(1,61)): # всего 60 страниц
    full_list.extend(get_tickers(p))
    
print(len(full_list))
```

```python
print(full_list)
```

```python
# Чистим от новых строк
clean_list = []
for x in full_list:
    clean_list.append(x.replace("\n", ""))       
print(clean_list)
```

```python
# Переводим с Pandas

import numpy as np
import pandas as pd

input = [x.strip(' ') for x in clean_list]

num = np.array(input)

reshaped = num.reshape(int(len(input)/3),3)

output = pd.DataFrame(reshaped, index=None, columns = ['#', "Ticker","MCap"])

output[2900:3000]
```

```python
output.to_csv('IXIC.csv', index=False)
```
